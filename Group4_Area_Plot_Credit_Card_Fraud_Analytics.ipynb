# Dataset Description

This dataset offers a variety of attributes valuable for comprehensive analysis. It contains 555,719 instances and 22 attributes, a mix of categorical and numerical data types. Importantly, the dataset is complete with no null values. Here's a breakdown of the attributes:

- **Trans_date_trans_time:** Timestamp of the transaction (date and time).
- **Cc_num:** Unique customer identification number.
- **Merchant:** The merchant involved in the transaction.
- **Category:** Transaction type (e.g., personal, childcare).
- **Amt:** Transaction amount.
- **First:** Cardholder's first name.
- **Last:** Cardholder's last name.
- **Gender:** Cardholder's gender.
- **Street:** Cardholder's street address.
- **City:** Cardholder's city of residence.
- **State:** Cardholder's state of residence.
- **Zip:** Cardholder's zip code.
- **Lat:** Latitude of cardholder's location.
- **Long:** Longitude of cardholder's location.
- **City_pop:** Population of the cardholder's city.
- **Job:** Cardholder's job title.
- **Dob:** Cardholder's date of birth.
- **Trans_num:** Unique transaction identifier.
- **Unix_time**: Transaction timestamp (Unix format).
- **Merch_lat**:Merchant's location (latitude).
- **Merch_long**: Merchant's location (longitude).
- **Is_fraud**:Fraudulent transaction indicator (1 = fraud, 0 = legitimate). This is the target variable for classification purposes.

# Fraud vs. Legitimate Transactions Over Time

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Grouping data by date for fraud and legitimate transactions separately and then plotting
daily_fraud = data_cleaned[data_cleaned['is_fraud'] == 1].resample('D', on='trans_date_trans_time')['amt'].sum()
daily_legitimate = data_cleaned[data_cleaned['is_fraud'] == 0].resample('D', on='trans_date_trans_time')['amt'].sum()

# Creating a dataframe for plotting
daily_transactions_comparison = pd.DataFrame({'Legitimate': daily_legitimate, 'Fraudulent': daily_fraud})
daily_transactions_comparison.fillna(0, inplace=True)  # Replacing NaNs with 0 for days without transactions

# Plotting area graphs
daily_transactions_comparison.plot(kind='area', stacked=False, alpha=0.5, figsize=(12, 6))

plt.title('Fraud vs. Legitimate Transactions Over Time (Daily Sum)')
plt.xlabel('Date')
plt.ylabel('Transaction Amount ($)')
plt.legend(loc='upper left')
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Converting the DataFrame to a long format suitable for Plotly plotting
daily_transactions_long = daily_transactions_comparison.reset_index().melt(id_vars=["trans_date_trans_time"], var_name="Transaction Type", value_name="Amount")

# Splitting the data by transaction type
legitimate_transactions = daily_transactions_long[daily_transactions_long['Transaction Type'] == 'Legitimate']
fraudulent_transactions = daily_transactions_long[daily_transactions_long['Transaction Type'] == 'Fraudulent']

# Creating the figure
fig = go.Figure()

# Adding Legitimate Transactions
fig.add_trace(go.Scatter(x=legitimate_transactions['trans_date_trans_time'],
                         y=legitimate_transactions['Amount'],
                         fill='tozeroy',
                         mode='lines',
                         name='Legitimate',
                         line_color='skyblue',
                         opacity=0.5))

# Adding Fraudulent Transactions
fig.add_trace(go.Scatter(x=fraudulent_transactions['trans_date_trans_time'],
                         y=fraudulent_transactions['Amount'],
                         fill='tozeroy',
                         mode='lines',
                         name='Fraudulent',
                         line_color='orange',
                         opacity=0.5))

# Updating the layout
fig.update_layout(title='Fraud vs. Legitimate Transactions Over Time (Daily Sum)',
                  xaxis_title='Date',
                  yaxis_title='Transaction Amount ($)',
                  legend_title='Transaction Type')

fig.show()
```

# Category-wise Fraud Transactions Over Time

```python
# we'll plot the daily sum of fraudulent transactions for each category
category_fraud_daily_non_cumulative = data_cleaned[data_cleaned['is_fraud'] == 1].groupby([pd.Grouper(key='trans_date_trans_time', freq='D'), 'category'])['amt'].sum().unstack(fill_value=0)

# Plotting area graph for each category
category_fraud_daily_non_cumulative.plot(kind='area', stacked=False, alpha=0.5, figsize=(14, 7))
plt.title('Category-wise Fraud Transactions Over Time (Daily Sum)')
plt.xlabel('Date')
plt.ylabel('Fraud Transaction Amount ($)')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# To focus on the top 2 categories affected by fraud the most, we'll first identify these categories by total fraud amount
top_categories = data_cleaned[data_cleaned['is_fraud'] == 1].groupby('category')['amt'].sum().nlargest(2).index

# Filtering the dataset for only these top categories
top_category_fraud_daily = data_cleaned[(data_cleaned['is_fraud'] == 1) & (data_cleaned['category'].isin(top_categories))].groupby([pd.Grouper(key='trans_date_trans_time', freq='D'), 'category'])['amt'].sum().unstack(fill_value=0)

# Define your custom colors for the top categories
# custom_colors = ['skyblue', 'salmon']

# Plotting area graph for top 3 categories
top_category_fraud_daily.plot(kind='area', stacked=False, alpha=0.5, figsize=(14, 7))
plt.title('Top 2 Categories Affected by Fraud Transactions Over Time (Daily Sum)')
plt.xlabel('Date')
plt.ylabel('Fraud Transaction Amount ($)')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Convert 'top_category_fraud_daily' DataFrame to a long format suitable for Plotly plotting
top_category_fraud_daily_long = top_category_fraud_daily.reset_index().melt(id_vars=["trans_date_trans_time"], var_name="Category", value_name="Fraud Amount")

# Define your custom colors for each category
custom_color_map = {
    'misc_net': 'lightblue',
    'shopping_net': 'orange'
}

# Using Plotly Express to create the area plot with custom colors
fig = px.area(top_category_fraud_daily_long,
              x="trans_date_trans_time",
              y="Fraud Amount",
              color="Category",
              title='Top 2 Categories Affected by Fraud Transactions Over Time (Daily Sum)',
              labels={"Fraud Amount": "Fraud Transaction Amount ($)", "trans_date_trans_time": "Date"},
              line_group="Category",
              color_discrete_map=custom_color_map)  # Apply custom colors here

# Adjusting the layout and transparency for clarity
fig.update_traces(opacity=0.5)  # Adjusting the opacity to make the area semi-transparent
fig.update_layout(legend_title='Category', legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1))

# Showing the plot
fig.show()
```

# Geographical Trends in Fraud Transactions oveer time

```python
# Aggregate fraud transactions by date and state
fraud_transactions_date_state = data_cleaned[data_cleaned['is_fraud'] == 1].groupby([pd.Grouper(key='trans_date_trans_time', freq='M'), 'state'])['amt'].sum().unstack(fill_value=0)

# Summing fraud amounts across all dates to identify top states
top_states_by_fraud = fraud_transactions_date_state.sum().nlargest(5).index

# Filtering the aggregated data for top states
fraud_transactions_top_states = fraud_transactions_date_state[top_states_by_fraud]

# # Plotting the area graph for fraud transaction amounts over time for top states
# fraud_transactions_top_states.plot(kind='area', stacked=True, alpha=0.5, figsize=(14, 7))
# plt.title('Fraud Transaction Amounts Over Time for Top States')
# plt.xlabel('Date')
# plt.ylabel('Fraud Transaction Amount ($)')
# plt.legend(title='State')
# plt.show()


# Plotting the area graph for fraud transaction amounts over time for top states
ax = fraud_transactions_top_states.plot(kind='area', stacked=True, alpha=0.5, figsize=(14, 7))
plt.title('Fraud Transaction Amounts Over Time for Top States')
plt.xlabel('Date')
plt.ylabel('Fraud Transaction Amount ($)')

# Setting the x-axis range
# Example: Setting the range from June 2020 to December 2020
ax.set_xlim(pd.Timestamp('2020-06-01'), pd.Timestamp('2020-12-31'))

plt.legend(title='State')
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Preparing the DataFrame for Plotly, which requires a long-form DataFrame
fraud_transactions_long = fraud_transactions_top_states.reset_index().melt(id_vars=["trans_date_trans_time"], var_name="State", value_name="Fraud Amount")

# Creating a stacked area chart with Plotly
fig = go.Figure()

for state in top_states_by_fraud:
    # Filter data for the current state
    state_data = fraud_transactions_long[fraud_transactions_long['State'] == state]
    fig.add_trace(go.Scatter(x=state_data['trans_date_trans_time'], y=state_data['Fraud Amount'], hoverinfo='x+y', mode='lines', name=state, stackgroup='one', fill='tonexty'))

# Updating layout for a clearer presentation
fig.update_layout(xaxis=dict(range=[fraud_transactions_long['trans_date_trans_time'].min(), fraud_transactions_long['trans_date_trans_time'].max()]),
                  title='Fraud Transaction Amounts Over Time for Top States', xaxis_title='Date', yaxis_title='Fraud Transaction Amount ($)', legend_title="State")
fig.show()
```
